```python
import requests
from bs4 import BeautifulSoup
url = "https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&oid=003&date=20210102"
response = requests.get(url,headers={'User-Agent':'Mozilla/5.0'})
soup = BeautifulSoup(response.text, 'html.parser')
li_list = soup.select("ul.type06_headline li")
for i in li_list:
    #a=i.select_one('a[href^="https://"]')
    #text="".join(a)
    art_url=i.find("a")["href"]
    response = requests.get(art_url,headers={'User-Agent':'Mozilla/5.0'})
    soup = BeautifulSoup(response.text, 'html.parser')
    title=soup.select('#content > div > div.content > div > div.news_headline > h4')
    content=soup.select('#newsEndContents')
    if title and content:
        print(f"{title[0].text}\n")
        print(content[0].text)
    else:
        continue
```

```python
import pickle
i = 0
URL = 'https://search.naver.com/search.naver?where=news&query=%EC%83%81%EC%97%85%EC%9A%A9%20%EB%B6%80%EB%8F%99%EC%82%B0%20%2B%EC%9E%84%EB%8C%80%EC%88%98%EC%9D%B5%20%2B%EC%9E%84%EB%8C%80%EB%A3%8C&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds=2020.01.01&de=2020.12.31&docid=&nso=so%3Ar%2Cp%3Afrom20200101to20201231%2Ca%3Aall&mynews=1&refresh_start='
while True: #조건을 모르겠음
    new_URL1 = URL + str((i-1)*10+1) + '&refresh_start=0'
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
    cookies = {'news_office_checked':'1009,1277,1015'}
    response = requests.get(new_URL1,headers=headers, cookies=cookies)
    soup = BeautifulSoup(response.text,"html.parser")
    li_list = soup.select('ul.list_news li')
    for li in li_list:
        go_naver = li.select_one('li.bx > div.news_wrap.api_ani_send > div > div.news_info > div > a:nth-child(3)')
        if go_naver:
            new_URL = go_naver.attrs['href']
            headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
            res = requests.get(new_URL,headers=headers)
            s = BeautifulSoup(res.text,"html.parser")
            td_list = s.select('#main_content')
            for td in td_list:
                title = td.select_one('div[class="article_info"]>#articleTitle').text
                body = td.select('#articleBodyContents')[0].text.replace('\n',"")
                date = td.select_one('div[class="sponsor"] span[class="t11"]').text
                press = td.select_one('div[class="press_logo"] img').attrs['title']
                print(title)
                print(press,date)
                print(body)
                print()
                with open('naver_news_property.csv','wb')as file:
                    pickle.dump(title,file)
                    pickle.dump(press,file)
                    pickle.dump(date,file)
                    pickle.dump(body,file)
                    
    i += 1
    if i > 4002: #(i-1)*10+1
        break
```
