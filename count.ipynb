```python
import numpy as np
import pandas as pd
from unicodedata import normalize
import datetime
```

```python
def daily_news_count(csv):
    news = pd.read_csv(f'{csv}')

    d = datetime.date(2020,1,1)
    t=d.strftime('%Y.%m.%d')

    while True:
        t=d.strftime('%Y.%m.%d')
        daily=news[news['date']==f'{t}']['body']
        count=0
        for i in daily:
            count+=1
        d = d + datetime.timedelta(days=1)
        print(t,count)
        if d == datetime.date(2021,1,1):
            break
```

# 메타 데이터 + 키워드 기사 카운트

```python
mf = pd.read_csv('Meta_Data_HK2.csv',index_col=0)
mf['count_keywords']='none'

d = datetime.date(2020,1,1)
t=d.strftime('%Y.%m.%d')

while True:
    t=d.strftime('%Y.%m.%d')
    daily=df[df['date']==f'{t}']['body']
    count=0
    for i in daily:
        count+=1
    d = d + datetime.timedelta(days=1)
    print(t,count)
    mf['count_keywords'][t]=count
    if d == datetime.date(2021,1,1):
        break
```

```python
mf['percentage'] = (mf['count_keywords']/mf['count']) * 100 # 백분율 계산
mf
```
