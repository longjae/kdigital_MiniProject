```python
import requests
from bs4 import BeautifulSoup
import datetime
import csv
import pandas as pd

import wordcloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from os import path
from collections import Counter
```

```python
def total_word_cloud(csv):
    tf = pd.read_csv(f"{csv}",index_col=0)
    text = tf['token']
#     text.split()

    # c = Counter(text)
    # print(c.most_common(20))

    # df_after['tokens'] = noun_words
    noun_text = ''
    FONT_PATH = 'C:/Windows/Fonts/malgun.ttf'
    for word in text:
        noun_text = noun_text + word
    # print(noun_text)
    wordcloud = WordCloud(max_font_size=200,
                          font_path=FONT_PATH,stopwords=STOPWORDS,
                            background_color='#FFFFFF',width=1200,
                            height=800).generate("".join(noun_text))
    plt.figure(figsize=(10,10))
    plt.imshow(wordcloud)
    plt.tight_layout(pad=0)
    plt.axis("off")
    plt.show()
```

```python
def total_word_count(csv):
    tf = pd.read_csv(f"{csv}",index_col=0)
    body = tf.loc['none']['token']
    text=body.split(",")

    c = Counter(text)
    print(c.most_common(10))
```
