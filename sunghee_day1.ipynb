```python
import requests
from bs4 import BeautifulSoup

# header를 사용해 방화벽(?)을 뚫음.
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
url ='https://finance.naver.com/item/frgn.nhn?code=330590&page=1'

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text, 'html.parser')

tr_list = soup.select('table.type2 > tr')

for tr in tr_list:
    if tr.select('td.num'):
        td_list=tr.select('span.tah')
        print(td_list[0].text, td_list[1].text, td_list[2].text, td_list[4].text)
```

```python
# 성희 님 방법

import requests
from bs4 import BeautifulSoup

# header를 사용해 방화벽(?)을 뚫음.
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
url ='https://finance.naver.com/item/frgn.nhn?code=330590&page=1'

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text, 'html.parser')

result = []

tr_list = soup.select('table.type2 > tr')

for tr in tr_list:
    if tr.select('td.num'):
        td_list=tr.select('span.tah')
        result.append([td_list[0].text,td_list[1].text,td_list[2].text,td_list[4].text])

result
```
